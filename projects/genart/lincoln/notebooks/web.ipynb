```python
class Trace:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.points = [[x, y]]
        
    
    def find_nearest_point(self, field):
        dists = []
        for p in field:
            dists.append(spatial.distance.euclidean([self.x, self.y], p))
            
        dists = np.array(dists)
        ind = np.where(dists == dists.min())[0][0]
        self.x = field[ind][0]
        self.y = field[ind][1]
        
        self.points.append([self.x, self.y])
        
        return np.delete(field, ind, 0)
    
class Node:
    def __init__(self, x, y, max_edges):
        self.x = x
        self.y = y
        self.max_edges = max_edges
        self.connections = 0
        
        def find_nearest_nodes(self, nodes):
            
            dists = []
            
            for node in nodes:
                if (node.x == self.x) & (node.y == self.y):
                    continue
                dists.append(spatial.distance.euclidean([self.x, self.y], [node.x, node.y]))

            dists = pd.Series(dists).sort_values()

            
            ind = dists.iloc[:max_edges].index
            
            
            field = np.delete(field, ind, 0)
        
        
```
